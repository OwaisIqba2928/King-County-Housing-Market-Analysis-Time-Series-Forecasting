### Step 1: Import Libraries

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset

```python
# Define the file path 
file_path = r"C:\Users\Asus\Music\Advance Analytics & Dashboard Design\kc_house_data.csv"
```

```python
# Load the dataset
df = pd.read_csv(file_path)
```

```python
# Display first 5 rows
df.head()
```

### Step 3: Check for Missing Values

```python
# Check for missing values
df.isnull().sum()
```

### Step 4: Check Data Types

```python
# Check data types
df.info()
```

### Step 5: Remove Duplicates

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

### Step 6: Summary Statistics

```python
# Display summary statistics
df.describe()
```

### Step 7: Count Unique Values in Each Column

```python
# Count unique values per column
df.nunique()
```

### Step 8: Distribution of House Prices

```python
# Plot histogram of prices
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=50, color='blue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

### Step 9: Bedrooms vs Price

```python
# Scatter plot of bedrooms vs price
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['bedrooms'], y=df['price'])
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms vs Price")
plt.show()
```

### Step 10: Square Footage vs Price

```python
# Scatter plot of sqft_living vs price
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['sqft_living'], y=df['price'])
plt.xlabel("Square Footage of Living Area")
plt.ylabel("Price")
plt.title("Living Space vs Price")
plt.show()
```

### Step 11: Average House Prices Per ZIP Code

```python
# Group by ZIP code and calculate mean price
zip_prices = df.groupby('zipcode')['price'].mean().sort_values(ascending=False)
```

```python
# Display top 10 most expensive ZIP codes
zip_prices.head(10)
```

### Step 12: Year Built vs Price

```python
# Scatter plot of year built vs price
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['yr_built'], y=df['price'])
plt.xlabel("Year Built")
plt.ylabel("Price")
plt.title("House Age vs Price")
plt.show()
```

### Step 13: Waterfront vs Non-Waterfront Prices

```python
# Average price comparison between waterfront and non-waterfront houses
df.groupby('waterfront')['price'].mean()
```
